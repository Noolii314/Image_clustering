```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
```

```python
kl = imread('graniti.jpg')
```

```python
plt.figure(figsize = (12,12))
plt.imshow(kl)
```

```python
kl.shape
```

```python
X = kl.reshape(960*720, 3)
```

```python
X.shape
```

```python
X
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 5)
km.fit(X)
```

```python
km.cluster_centers_
```

```python
img_seg = km.cluster_centers_
```

```python
km.labels_
```

```python
img_seg = img_seg[km.labels_]
```

```python
img_seg.shape
```

```python
img_seg = img_seg.reshape(kl.shape)
```

```python
plt.figure(figsize = (12,12))
plt.imshow((img_seg ).astype(np.uint8))
```
